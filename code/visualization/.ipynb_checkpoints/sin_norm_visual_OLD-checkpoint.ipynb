**REMVING OUTLIERS TO PLOT**

```python
#Removing outliers to print diagrams: 
df_sinout = df.copy()
for S in df_sinout.columns:
    if df_sinout[S].dtypes=='float64':        
        df_sinout=df_sinout[~((df_sinout[S]-df_sinout[S].mean()).abs() > 3*df_sinout[S].std())]
print(df_sinout.shape)
print(df.shape)

dfdensity=df_sinout.iloc[:,1:]
names=dfdensity.iloc[:,:-1].columns
```
