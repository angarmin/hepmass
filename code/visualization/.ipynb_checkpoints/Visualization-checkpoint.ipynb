```python
print("CON ZOOM, ESCALADO MANUAL")

cols=['jet1_pt','jet2_pt', 'jet3_pt', 'jet4_pt', 'lep_pt', 'met_miss'
      , 'jet1_eta', 'jet1_phi', 'jet2_eta', 'jet2_phi',
       'jet3_eta', 'jet3_phi', 'jet4_eta', 'jet4_phi', 'lep_eta',
       'lep_phi', 'met_phi', 'jets_no', 'm_jj',
       'm_jjj', 'm_lv', 'm_jlv', 'm_wwbb']

xlabel=['GEvs','GEvs', 'GEvs', 'GEvs', 'GEvs', 'GEvs'
      , 'adimensional', 'rads', 'adimensional', 'rads',
       'adimensional', 'rads', 'adimensional', 'rads', 'adimensional',
       'rads', 'rads', 'Número de jets', 'GEvs',
       'GEvs', 'GEvs', 'GEvs', 'GEvs']

dfdensity=df.copy()


for names in cols: 
    if names == "lep_pt":
        min=0
        max=400
    elif names=='met_miss':
        min=0
        max=600
    elif names=='jet1_pt':
        min=0
        max=1000
    elif names=='jet2_pt':
        min=0
        max=600
    elif names=='jet3_pt':
        min=0
        max=300
    elif names=='jet4_pt':
        min=0
        max=200
    elif names == "m_jj":
        min=0
        max=250
    elif names=='m_jjj':
        min=120
        max=220
    elif names=='m_lv':
        min=75
        max=85
    elif names=='m_jlv':
        min=0
        max=1100
    elif names=='m_wwbb':
        min=0
        max=4000
        
    elif names=='jets_no':
        min=0
        max=0
        
    elif names=='jets_no':
        min=0
        max=0
 
    else: 
        min =df[names].min()
        max =df[names].max()


    SIZE=6
    fig, ax = plt.subplots(SIZE, 1,figsize=(10,5))

    for i in range(SIZE):
        if i !=0:
            mass=250+250*i
        else:
            mass=0
            #plt.title(names)
        data1=dfdensity[dfdensity.mass==mass][names]
        
        if  names == "jets_no":
            ax[i].bar(range(4,14), data1.value_counts()/data1.shape[0] , color = 'blue', edgecolor = 'black', alpha=0.3)
        else:
        
            ax[i].hist(data1,color = 'blue', edgecolor = 'black', density=True, bins=50,range=[min,max],alpha=0.3)
        ax[i].set_yticks([]) 
        ax[i].tick_params(axis="x", labelsize=14)

        ax[i].set_ylabel(mass,fontsize=15)
       
    fig.suptitle(names,fontsize=16)
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.xlabel(xlabel[cols.index(names)],fontsize=15)
    plt.show()
```

** DIFFERENCE BETWEEN BTAG CORRECT AND UNCORRECT** 

```python
print("CON ZOOM, ESCALADO MANUAL")
cols=['jet1_pt','jet2_pt', 'jet3_pt', 'jet4_pt', 'lep_pt', 'met_miss'
      , 'jet1_eta', 'jet1_phi', 'jet2_eta', 'jet2_phi',
       'jet3_eta', 'jet3_phi', 'jet4_eta', 'jet4_phi', 'lep_eta',
       'lep_phi', 'met_phi', 'jets_no', 'm_jj',
       'm_jjj', 'm_lv', 'm_jlv', 'm_wwbb']

xlabel=['GEvs','GEvs', 'GEvs', 'GEvs', 'GEvs', 'GEvs'
      , 'adimensional', 'rads', 'adimensional', 'rads',
       'adimensional', 'rads', 'adimensional', 'rads', 'adimensional',
       'rads', 'rads', 'Número de jets', 'GEvs',
       'GEvs', 'GEvs', 'GEvs', 'GEvs']





dfdensity0=df[df.loc[:,['jet1_btag','jet2_btag',"jet3_btag","jet4_btag"]].sum(axis=1)==0]
dfdensity1=df[df.loc[:,['jet1_btag','jet2_btag',"jet3_btag","jet4_btag"]].sum(axis=1)==1]
dfdensity2=df[df.loc[:,['jet1_btag','jet2_btag',"jet3_btag","jet4_btag"]].sum(axis=1)==2]


for names in cols: 
    if names == "lep_pt":
        min=0
        max=400
    elif names=='met_miss':
        min=0
        max=600
    elif names=='jet1_pt':
        min=0
        max=1000
    elif names=='jet2_pt':
        min=0
        max=600
    elif names=='jet3_pt':
        min=0
        max=300
    elif names=='jet4_pt':
        min=0
        max=200
    elif names == "m_jj":
        min=0
        max=250
    elif names=='m_jjj':
        min=120
        max=220
    elif names=='m_lv':
        min=75
        max=85
    elif names=='m_jlv':
        min=0
        max=1100
    elif names=='m_wwbb':
        min=0
        max=4000

    elif names=='jets_no':
        min=0
        max=0

    else: 
        min =df[names].min()
        max =df[names].max()

    SIZE=6
    fig, ax = plt.subplots(SIZE, 1,figsize=(20,8))

    for i in range(SIZE):
        if i !=0:
            mass=250+250*i
        else:
            mass=0
            #plt.title(names)
        data0=dfdensity0[dfdensity0.mass==mass][names]

        if  names == "jets_no":
            ax[i].bar(range(4,14), data0.value_counts()/data0.shape[0] , color = 'blue', edgecolor = 'black', alpha=0.3,label=("n b-jets =" + str(0)))
        else:

            ax[i].hist(data0,color = 'blue', edgecolor = 'black', density=True, bins=50,range=[min,max],alpha=0.3,label=("n b-jets =" + str(0)))         
        
        ax[i].set_yticks([]) 
        ax[i].set_ylabel(mass,fontsize=15)
        ax[i].tick_params(axis="x", labelsize=14)
    ax[0].legend(loc="upper right")
        
    for i in range(SIZE):
        if i !=0:
            mass=250+250*i
        else:
            mass=0

        data1=dfdensity1[dfdensity1.mass==mass][names]

        if  names == "jets_no":
            ax[i].bar(range(4,14), data1.value_counts()/data1.shape[0] , color = 'green', edgecolor = 'black', alpha=0.3,label=("n b-jets =" + str(1)))
        else:

            ax[i].hist(data1,color = 'green', edgecolor = 'black',density=True, bins=50,range=[min,max],alpha=0.3,label=("n b-jets =" + str(1)))
        ax[i].set_yticks([]) 
        ax[i].set_ylabel(mass,fontsize=15)
        ax[i].tick_params(axis="x", labelsize=14)
    ax[0].legend(loc="upper right")

    for i in range(SIZE):
        if i !=0:
            mass=250+250*i
        else:
            mass=0

        data2=dfdensity2[dfdensity2.mass==mass][names]

        if  names == "jets_no":
            ax[i].bar(range(4,14), data2.value_counts()/data2.shape[0] , color = 'red', edgecolor = 'black', alpha=0.3,label=("n b-jets =" + str(2)))
        else:

            ax[i].hist(data2,color = 'red', edgecolor = 'black',density=True, bins=50,range=[min,max],alpha=0.3,label=("n b-jets =" + str(2)))
        ax[i].set_yticks([]) 
        ax[i].set_ylabel(mass,fontsize=15)
        ax[i].tick_params(axis="x", labelsize=14)
    ax[0].legend(loc="upper right",fontsize=12)      
        




    fig.suptitle(names,fontsize=16)
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.xlabel(xlabel[cols.index(names)],fontsize=15)
    plt.show()
```

## Correlation Plot

```python
df_corr=df[['lep_pt', 'met_miss', 
       'jet1_pt', 'jet2_pt',
       'jet3_pt', 'jet4_pt', 'jets_no','m_jj',
       'm_jjj', 'm_lv', 'm_jlv', 'm_wwbb', 'mass','jets_no']].sample(n=10000, random_state=1)


df_corr=df_corr.corr() 
    
```

```python
df_corr
```

```python

print("SCORRELATION PLOT FOR NORMALIZED DATA (7 MILLIONS)")


plt.figure(figsize=(15,15))

corr_sin_norm=df_corr.round(2)
#g=sns.heatmap(corr_sin_norm, mask=np.zeros_like(df_corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            #square=True, annot = True, annot_kws = {"size" : 10},cbar=False)

g=sns.heatmap(corr_sin_norm, mask=np.zeros_like(corr_sin_norm, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
           square=True, annot = True, annot_kws = {"size" : 10},cbar=False)


plt.title('Matriz de correlacion lineal entre variables', fontsize=20)
plt.yticks(rotation=45, fontsize=15) 
plt.xticks(rotation=45, fontsize=15) 


bottom, top = g.get_ylim()
g.set_ylim(bottom + 0.5, top - 0.5)

plt.show()


fig = g.get_figure()
fig.savefig(pwdsnorm  + 'correlation.png')    
#Keep in mind: categorical as number of jets are not HERE!

```
